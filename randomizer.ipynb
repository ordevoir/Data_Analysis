```python
import random

class RandomWithoutRepeat:
    def __init__(self, start=1, end=16):
        self.start = start
        self.end = end
        self.reset()
    
    def reset(self):
        self.numbers = list(range(self.start, self.end + 1))
        random.shuffle(self.numbers)
        self.index = 0
    
    def next(self):
        if self.index >= len(self.numbers):
            print("Autoreset")
            self.reset()
        result = self.numbers[self.index]
        self.index += 1
        return result

rng = RandomWithoutRepeat(1, 16)
```
